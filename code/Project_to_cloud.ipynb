```python
def correlationt(dataset, threshold):
    X_cols=set()
    X_corr=dataset.corr()
    for i in range(len(X_corr.columns)):
        for j in range(i):
            if abs(X_corr.iloc[i,j]) > threshold:                
                X_cols.add(X_corr.index[i])                      
    return(X_cols)
```

```python
correlationt(X,0.85)
```

```python
X.drop(columns=correlationt(X,0.85),inplace=True)
```
